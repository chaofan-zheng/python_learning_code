```python
def convert_currency(var):
    new_value = var.replace("平米","")
    return new_value
```

```python
data.apply(convert_currency(data['square']))
```

```python
#得到最小平米数和最大平米数
def pingmi(data,b):
    if b in data:
        return float(data[0:data.find(b)])
    else:
        return None
```

```python
data['new_square']=data['square'].apply(pingmi,b='平米')
data['new_square']
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
#  数据分析1：找出所有的户型种类及价格最高的5个别墅
#找出所有户型种类
data.housetype.value_counts().head(10).plot(kind='barh')
```

```python
#价格最高的五栋别墅
bieshu=data[data.housetype.str.contains('别墅')]
bieshu.sort_values(by='totalPrice',ascending=False).head(5)
```

```python
#数据分析3：找出关注人数最多的五套房子
#提取出关注人数
data['guanzhu']=data['followInfo'].apply(pingmi,b='人关注')
data['guanzhu']
data.head(1)
```

```python
#找出关注最多的五套房
data.guanzhu=pd.to_numeric(data.guanzhu, errors='coerce')
top_5_guanzhu=data.sort_values(by='guanzhu',ascending=False).head(5)
top_5_guanzhu
```

```python
#据分析4：户型和关注人数分布绘制条形图进行可视化
```

```python
housetype_guanzhu=data['guanzhu'].groupby(data['housetype'])
housetype_guanzhu.count().head(10).sort_values(ascending=True).plot(kind='barh')
```

```python
#  数据分析5：面积分布
bins=[0,50,100,150,200,250,300,350,400,450,500]
data['new_square1']=pd.cut(data.new_square,bins=bins)
data['new_square1'].value_counts().sort_values(ascending=True).plot(kind='barh')
```

```python
#数据分析6：各个行政区房源均价，柱形图绘制，箱线图绘制，在售房源信息统计
new_data=pd.merge(data,data1,how='left', on='community')
new_data.head(10)
```

```python
district_mean=new_data['totalPrice'].groupby(new_data['district'])
district_mean.mean().sort_values(ascending=True).plot(kind='barh')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x='district',y='totalPrice',data=new_data)
plt.ylim((0,2000))
plt.title('箱线图绘制各行政区房价情况')
```

```python
#数据分析8：按照地铁信息对各个区域均价排序，柱形图绘制
new_data['bizcircle'].dropna(axis=0)
subway=new_data['totalPrice'].groupby(new_data['bizcircle']).mean()
subway.sort_values(ascending=True).head(10).plot(kind='barh')
```

```python
#数据分析9：按小区均价排序，柱形图绘制
community=new_data['totalPrice'].groupby(new_data['community']).mean()
community.sort_values().head(10).plot(kind='barh')
```

```python
# 将字符串转换成数字
def data_ads(select_data,str):
    if str in select_data:
       return (select_data[0:select_data.find(str)])
    else:
       return '没有提取到楼层信息'
# 得到楼层
# 将字符串转换成数字
new_data['layer']=new_data['years'].apply(data_ads,str='(')
new_data.head(3)
```

```python
list2=[]
for i in new_data['new_floor']:
    list2.append(i.replace("(","a"))
list2
```

```python
new_data['new_floor']=list2
```

```python
def pingmi1(data,b):
    if b in data:
        return (data[0:data.find(b)])
    else:
        return None
```

```python
new_data['new_floor2']=new_data['new_floor'].apply(pingmi1,b='a')
new_data['new_floor2']
```

```python
new_data['new_floor2'].value_counts().plot(kind='barh')
```

```python
#数据分析11：绘制2000到2016平均房价
list3=[]
for i in list2:
    list3.append(i.replace(")","b"))
new_data['new_years']=list3
new_data['new_years']
```

```python
def nian(data,b):
    if b in data:
        return(data[data.find(b)-5:data.find(b)])
```

```python
a=new_data['new_years'].apply(nian,b='建')
a.dropna(inplace=True)
```

```python
new_data['year']=a.apply(pingmi,b='年')
```

```python
new_data['year'].value_counts().head(10).plot(kind='barh')
```
