# 创建数据透视表
对两组数据进行分组，对某一个值进行聚合操作

```python
data_gender2=pd.pivot_table(data=movielens,index='Title',columns='Gender',values='Rating',aggfunc='mean')
data_gender2
```

```python
# 男女评分分歧最大的五部电影
# 新建一列['diff']
data_gender2['diff']=np.fabs(data_gender2.F-data_gender2.M)
data_gender2_sorted=data_gender2.sort_values(by='diff',ascending=False)
data_gender2_sorted.head(10)
```

```python
# 得到评分最高的十部电影(过滤掉点评次数少于100次的电影)
movies_stats=movielens.groupby('Title').agg({'Rating':['mean','size']})
movies_stats.head()
```

```python
atleast_100=movies_stats['Rating']['size']>=100
atleast_100.head()
```

```python
# 按照Rating里面的mean排序
movies_stats[movies_stats['Rating']['size']>=100].sort_values(by=('Rating','mean'),ascending=False)
```
