## 纵向合并

```python
import numpy as np
import pandas as pd
df1=pd.DataFrame(data=np.ones((3,4))*0,columns=list('abcd'))
df2=pd.DataFrame(data=np.ones((3,4))*1,columns=list('abcd'))
df3=pd.DataFrame(data=np.ones((3,4))*2,columns=list('abcd'))
print(df1)
print(df2)
print(df3)
```

```python
res=pd.concat(objs=[df1,df2,df3],axis=0,ignore_index=True) # ingnore_index重置行索引
res
```

```python
df1=pd.DataFrame(data=np.ones((3,4))*0,columns=list('abcd'),index=['1','2','3'])
df2=pd.DataFrame(data=np.ones((3,4))*1,columns=list('bcde'),index=['2','3','4'])
```

```python
res=pd.concat(objs=[df1,df2],axis=0,join='inner')
res
```

```python
# 构建层次索引
res=pd.concat(objs=[df1,df2],axis=0,join='inner',
             keys=['df1','df2'],
             names=['df_name','row_id'])
res
```

## 横向合并

```python
res=pd.concat(objs=[df1,df2],axis=1,join='inner',
             keys=['df1','df2'],
             names=['df_name','columns_id'])
res
```

```python
res=pd.concat(objs=[df1,df2],axis=1,join='outer',sort=True)
res
```

## append方法进行完全列相同的表的纵向合并

```python
df1=pd.DataFrame(data=np.ones((3,4))*0,columns=list('abcd'))
df2=pd.DataFrame(data=np.ones((3,4))*1,columns=list('abcd'))
df1.append(df2)
```
