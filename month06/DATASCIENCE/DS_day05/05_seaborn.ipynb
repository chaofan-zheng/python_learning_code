```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
# 忽略警告
import warnings
warnings.filterwarnings('ignore')
data=pd.read_csv('../DS_day03/data/iris.csv')
data.head()
sns.set_style('darkgrid')
```

```python
data.info()
```

```python
data.describe().T
```

```python
# distplot的使用
np.random.seed(123)
# 标准正态分布
data01=np.random.randn(10000)
sns.distplot(data01)
# 服从µ=loc,ø=scale的正态分布
data02=np.random.normal(loc=2,scale=3,size=10000)
sns.distplot(data02)
```

```python
# 服从µ=loc,ø=scale的正态分布
data02=np.random.normal(loc=6,scale=3,size=10000)
sns.distplot(data02)
```

```python
sns.distplot(data01,kde=False,bins=10)
plt.xticks([-3,-2,-1,0,1,2,3])
plt.xlabel('random')
```

```python
# 语文成绩
grade=np.random.normal(loc=70,scale=10,size=1000)
sns.distplot(grade)
```

# 鸢尾花listplot调查花瓣长度的分布

```python
sns.distplot(data['petal_length'],vertical=True,rug=True) # rug就是能把数值显示出来
```

```python
sns.distplot(data['petal_length'],kde=False)
```

# Boxplot 调查

```python
sns.boxplot(x='petal_length',data=data)# 需要data参数，distplot是不需要的
```

```python
sns.boxplot(y='petal_length',data=data) # x,y代表方向
```

```python
sns.boxplot(x='species',y='petal_length',data=data)
```

# countplot 先计数，后绘制

```python
# 对不同类别的数据进行计数
# pandas
data['species'].value_counts().plot(kind='bar')
```

```python
# seborn
sns.countplot(x='species',data=data)
```

```python
# barplot
import numpy as np
sns.barplot(x='species',y='petal_length',
           data=data,estimator=np.mean) # estimator 指定的是聚合的方法 ci=None可以去除误差线，ci指的是置信区间的大小
```
