# 二. 对于数据做描述性统计分析
主要内容：
1. 得到数据的形状，和数据的完成性
2. 熟悉数据中的特征
    - 数值型：求和，四分位数，标准差，filter差等 
        - 可视化：折线图，柱状图，密度曲线图，箱线图
    - 类别型数据，频数统计 
        - 可视化：柱状图

```python
# 默认情况下：得到数值型数据的统计量
# titanic.describe()
titanic.describe().T  # 转置
```

```python
# 类别型的数据统计
# count 非空元素的数目 unique类别 top数目最多的类别 freq 数目最多的类别的frequency
titanic.select_dtypes('object').describe().T
```

```python
# 类别型的详细数据
print(titanic['Sex'].value_counts())
print(titanic['Embarked'].value_counts()) # 类别的详细的值分别的统计
# value_counts也是可以统计数据的
print(titanic['Pclass'].value_counts())
# value_counts 是series的一个函数
```
