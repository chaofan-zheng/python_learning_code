# 读取文件
- filepath
- sep
- header
- names
- index_col
- dtype
- engine 默认是c，要改成python
- nrows

```python
import pandas as pd
import numpy as np
```

```python
#显示所有列
pd.set_option('display.max_columns', None)

#显示所有行
pd.set_option('display.max_rows', None)

#设置value的显示长度为100，默认为50
pd.set_option('max_colwidth',100)
```

```python
# 读取电影详情表
movies = pd.read_table(filepath_or_buffer='data/movies.dat',
                      sep='::',engine='python',
                      header=None, # 将某行数据作为列名，
                      names=['MovieID','Title','Genres'],) # 设定列名
                      #index_col=0) # 第0列作为行索引
movies.head()
```

```python
# 读取用户表
users = pd.read_table(filepath_or_buffer='data/users.dat',
                     sep='::',engine='python',
                     header=None,names=['UserID','Gender','Age','Occupation','Zip-code'])
users
```

```python
# 读取评分数据
ratings = pd.read_table(filepath_or_buffer='data/ratings.dat',
                     sep='::',engine='python',
                     header=None,names=['UserID','MovieID','Rating','Timestamp'])
ratings.head()
```

# 合并

```python
print(users.shape)
print(movies.shape)
print(ratings.shape)
```

```python
# 先把用户表和评分表合并，然后再跟movies合并
data = pd.merge(pd.merge(users,ratings),movies)
data.tail()
```
